```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata=pd.read_csv('Food Prices.csv')
mydata
```

```python
mydata
```

```python
dat=mydata[mydata['Year ']==2022]
dat
```

```python
dat[dat['Country ']=='Sweden']
```

```python
dat[dat['Country ']=='Sweden'].count()
```

```python
swe_dat=dat[dat['Country ']=='Sweden']#Food price analysis of sweden in fiscal year 2022
swe_dat
```

```python
print("Price of bread ðŸ¥ª in Sweden in succesive months of year 2022......")
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="Month ",y="Average Price ",data=swe_dat[swe_dat['Food Item ']=='Bread'],hue='Month ')
plt.show()
```

```python
print("Price analysis of Milk ðŸ¥›  in the year 2022 in Sweden")
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="Month ",y="Average Price ",data=swe_dat[swe_dat['Food Item ']=='Milk'])
plt.show()
```

```python
print("Price analysis of Potato in the year 2022 in Sweden.")
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="Month ",y="Average Price ",data=swe_dat[swe_dat['Food Item ']=='Potatoes'],hue='Month ')
plt.show()
```
